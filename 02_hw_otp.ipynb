```python
# Вновь разделите на train и test (не мешая)
# непонятно что именно разбить
new_train = pd.read_csv('02_otp_train.csv', sep=';')
new_train['PERSONAL_INCOME'] = new_train['PERSONAL_INCOME'].map(lambda x: x.replace(',', '.')).astype('float')
new_train['CREDIT'] = new_train['CREDIT'].map(lambda x: x.replace(',', '.')).astype('float')
new_train['FST_PAYMENT'] = new_train['FST_PAYMENT'].map(lambda x: x.replace(',', '.')).astype('float')
new_train['LOAN_AVG_DLQ_AMT'] = new_train['LOAN_AVG_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
new_train['LOAN_MAX_DLQ_AMT'] = new_train['LOAN_MAX_DLQ_AMT'].map(lambda x: x.replace(',', '.')).astype('float')
new_train= new_train.apply(lambda x: pd.to_numeric(x, errors='coerce')).dropna(axis=1) 
new_train
new_y = new_train['TARGET']
new_train = new_train.drop('TARGET', axis=1)

# Ваш код :-)
```

```python
# Сделать hold-out sampling на train выборке
from sklearn.cross_validation import train_test_split

# test_size=0.3, random_state=42
X_train, X_test, y_train, y_test = train_test_split(new_train, new_y, test_size=0.2, random_state=42)
# Ваш код :-)
```

```python
# Попробовать следующие "черные ящики": интерфейс одинаковый 
#     fit, 
#     predict, 
#     predict_proba

from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
dtc_predict = dtc.predict(X_test)
dtc_predict2 = dtc.predict_proba(X_test)

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc_predict = rfc.predict(X_test)

lr = LogisticRegression()
lr.fit(X_train, y_train)
lr_predict = rfc.predict(X_test)

# Ваш код :-)
```

```python
# Посчитать метрики стандартные
# accuracy, precision, recall
from sklearn.metrics import precision_score, recall_score, accuracy_score, f1_score

#DecisionTreeClassifier
print('precision', precision_score(y_test, dtc_predict))
print('recall', recall_score(y_test, dtc_predict))
print('accuracy', accuracy_score(y_test, dtc_predict))
print()

#RandomForestClassifier
print('precision', precision_score(y_test, rfc_predict))
print('recall', recall_score(y_test, rfc_predict))
print('accuracy', accuracy_score(y_test, rfc_predict))
print()

#LogisticRegression
print('precision', precision_score(y_test, lr_predict))
print('recall', recall_score(y_test, lr_predict))
print('accuracy', accuracy_score(y_test, lr_predict))
print()

# Ваш код :-)
```

```python
# Визуалищировать эти метрики всех моделей на одном графике (чтоб визуально посмотреть)
# визуализировать числа? 

# Ваш код :-)
```

```python
# Потроить roc-кривые всех можелей на одном графике
# Вывести roc_auc каждой моделе
%matplotlib inline
from matplotlib import pyplot as plt
plt.figure(figsize=(5, 5))

from sklearn.metrics import roc_auc_score, roc_curve

dtc_predict = dtc.predict_proba(X_test)
fpr_d, tpr_d, thresholds_d = roc_curve(y_test, dtc_predict[:,1])
p_dtc, = plt.plot(fpr_d, tpr_d, label='DecisionTreeClassifier')

rfc_predict = rfc.predict_proba(X_test)
fpr_r, tpr_r, thresholds_r = roc_curve(y_test, rfc_predict[:,1])
p_rfc, = plt.plot(fpr_r, tpr_r, label='RandomForestClassifier')

lr_predict = lr.predict_proba(X_test)
fpr_l, tpr_l, thresholds_l = roc_curve(y_test, lr_predict[:,1])
p_lg, = plt.plot(fpr_l, tpr_l, label='LogisticRegression')

plt.legend(handles=[p_dtc, p_rfc, p_lg])

plt.plot([0, 1], [0, 1])
plt.ylabel('tpr')
plt.xlabel('fpr')
plt.grid(True)
plt.title('ROC curve')
plt.xlim((-0.01, 1.01))
plt.ylim((-0.01, 1.01))
# Ваш код :-)
```

```python
from sklearn.cross_validation import cross_val_score
# Сделать k-fold (10 фолдов) кросс-валидацию каждой модели
# И посчитать средний roc_auc

print(cross_val_score(dtc, new_train, new_y, scoring='roc_auc', cv=10).mean())

print(cross_val_score(rfc, new_train, new_y, scoring='roc_auc', cv=10).mean())

print(cross_val_score(lr, new_train, new_y, scoring='roc_auc', cv=10).mean())

# Ваш код :-)
```
